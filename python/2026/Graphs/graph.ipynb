```python
#no of vertices
n=8
A=[[0,1],[1,2],[3,1],[3,7],[5,1],[5,2]]

print(A)
```

```python
#adjacency matrix
M=[]
for i in range(n):
    M.append([0]*n)
M
"""
matrix = [[0]*n for _ in range(n)]
matrix
"""
```

```python
for u,v in A:
    M[u][v]=1
M
```

```python
#adjacency list
d={}
for u,v in A:
    if u not in d:
        d[u]=[]
    if v not in d:
        d[v]=[]
    d[u].append(v)
    d[v].append(u)
d
```

```python
#using class  for Adjacency matrix
class Graph:

    def __init__(self,n,directed=False):
        self.n=n
        self.directed=directed

        m=[]
        for i in range(self.n):
            m.append([0]*self.n)
        self.m=m
    
    def add_edge(self,u,v):
        self.m[u][v]=1
        if self.directed==False:
            self.m[v][u]=1
    def display(self):
        for i in self.m:
            print(i)

#Vertices
n=5 
a_m=Graph(n,directed=False) # a_m=Graph(n,directed=True)
a_m.add_edge(1,2)
a_m.add_edge(2,4)
a_m.add_edge(3,4)
a_m.display()
```

```python
class Graphs:
    def __init__(self,n,directed=False):
        self.n=n
        self.directed=directed

        m=[]
        for i in range(self.n):
            m.append([])
        self.m=m
    
    def add_edge(self,u,v):
        self.m[u].append(v)

        if self.directed==False:
            self.m[v].append(u)
    def display(self):
        for i in range(self.n):
            print(i,">>",self.m[i])

# Vertices
n = 5
a_l = Graphs(n, directed=False)   # a_l = Graph(n, directed=True)

a_l.add_edge(1,2)
a_l.add_edge(4,1)
a_l.add_edge(3,4)

a_l.display()
```
